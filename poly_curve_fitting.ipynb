```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures as pf 
from sklearn.linear_model import LinearRegression as lr 
import math as m 
x = np.array([m.sin(22*i/42) for i in range(13)])
print(x)
y = np.array([0.1,0.41,0.94,1.2,0.78,0.45,-0.12,-0.57,-0.87,-0.98,-0.88,-0.50,0.0])
```

```python
#  ploting of x
plt.Figure()
plt.title("original values of sin function (x)")
plt.plot([i for i in range(13)],x,color="blue")
plt.grid()
plt.scatter([i for i in range(13)],x)
```

```python
#ploting of y 
plt.figure()
plt.plot([t for t in range(13)],y,color = "green")
plt.grid()
plt.title("training set")
plt.scatter([i for i in range(13)],y)
```

```python
#combining the training and the original sets
plt.figure()
plt.plot([i for i in range(13)],x)
plt.plot([i for i in range(13)],y)
plt.grid()
```

```python
#training the sets
model = lr()
error = np.array([])
for i in range(20):
    poly  = pf(degree=i,include_bias=True)
    y_poly = poly.fit_transform(y.reshape(-1,1))
    model.fit(y_poly,x)
    rms_values = np.mean((model.predict(y_poly)-x)**2)
    error= np.append(error,rms_values)
    if(i==18):
        plt.Figure()
        plt.plot([i for i in range(13)],model.predict(y_poly))
        plt.plot([i for i in range(13)],x)
        plt.scatter([i for i in range(13)],model.predict(y_poly))
        plt.title("trained the model final output")
        plt.grid()
index = np.where(error == np.min(error))
print(error[index[0][0]])
print(index[0][0])
plt.figure()
plt.plot([i for i in range(13)],x)
plt.grid()
plt.title("original values")
plt.scatter([i for i in range(13)],x)
```

```python
plt.Figure()
plt.plot(range(20),error,marker = "o")
plt.xlabel("polynomial degree")
plt.ylabel("mean squared degree")
plt.title("error vs polynomial degree")
plt.grid()
```
