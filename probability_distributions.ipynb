```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import norm
mean =0
sigma = 1
x = np.linspace(-4,4,1000)
pdf = norm.pdf(x,mean,sigma)
plt.plot(x,pdf,label="normal distribution")
plt.plot([0*i for i in range(5)],[0,0.1,0.2,0.3,0.4])
plt.title("normal distribution(mean =0 sigma =1)")
plt.xlabel("x")
plt.ylabel("probability densitiy")
plt.legend()
plt.grid()
plt.show()
```
